```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("bike_sales_india.csv")
```

```python
print(df.shape)
df.head()
```

```python
df.info()
print(f"Total number of null values present in the dataset are {kawasaki.isnull().sum().sum()}")
```

```python
df["Brand"].value_counts()
```

```python
kawasaki=df[df["Brand"]=="Kawasaki"]
```

```python
data=kawasaki.groupby("Model")["Price (INR)"].sum()
top_brand = data.idxmax()
plt.figure(figsize=(6,4))
plt.bar(data.index, data.values)
plt.title(f"{top_brand} Model of Kawasaki Generated the Highest Total Sales Value")
plt.xlabel("Model")
plt.ylabel("Total Sales Value (INR)")   
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

plt.hist(kawasaki["Price (INR)"], bins=20, edgecolor="black")
plt.title("Uniform Distribution of Kawasaki Bike Prices")
plt.xlabel("Price (INR)")
plt.ylabel("Number of Bikes")
plt.show()
```

```python
data=kawasaki.groupby("Model")["Price (INR)"].sum()
top_brand = data.idxmax()
plt.bar(data.index, data.values)
plt.title(f"The Kawasaki {top_brand} model generated the highest total sales value")
plt.xlabel("Brand")
plt.ylabel("Total Sales Value (INR)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data=kawasaki.groupby("Model")["Price (INR)"].mean()
top_brand = data.idxmax()
plt.bar(data.index, data.values)
plt.title(f"The Kawasaki {top_brand} model generated the highest average revenue")
plt.xlabel("Brand")
plt.ylabel("Total Sales Value (INR)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data=kawasaki.groupby("State")["Price (INR)"].sum().sort_values(ascending=True)
top_state = data.idxmax()
plt.barh(data.index, data.values)
plt.title(f"{top_state} generates highest total sales value from Kawasaki bikes ")
plt.xlabel("State")
plt.xlabel("Total Sales Value (INR)")   
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data=kawasaki.groupby("State")["Price (INR)"].mean().sort_values(ascending=True)
top_state = data.idxmax()
plt.barh(data.index, data.values)
plt.title(f"{top_state} generates highest average revenue from Kawasaki bikes")
plt.xlabel("State")
plt.xlabel("Total Sales Value (INR)")   
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data=kawasaki.groupby("Year of Manufacture")["Price (INR)"].sum()
top_year = data.idxmax()
plt.plot(data.index.astype(str), data.values, marker="o", color="red", linewidth=3)
plt.title(f"{top_year} recorded the highest sales value of Kawasaki bikes")
plt.xlabel("Year of Manfacture")
plt.ylabel("Total Sales Value (INR)")   
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data=kawasaki.groupby("Registration Year")["Model"].count()
top = data.idxmax()
plt.bar(data.index, data.values)
plt.title(f"Kawasaki bikes had the highest number of registrations in {top} ")
plt.xlabel("Brand")
plt.ylabel("Total Sales Value (INR)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data=kawasaki["Owner Type"].value_counts()
top_owner = data.idxmax()
plt.bar(data.index, data.values)
plt.title(f"{top_owner}-owner bikes make up the highest number of Kawasaki bikes")
plt.xlabel("Owner Type")
plt.ylabel("Number of Bikes")
plt.tight_layout()
plt.show()
```

```python
kawasaki["Fuel Type"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90 )
plt.title("Petrol-powered Kawasaki models are widely used")
plt.show()
```

```python
kawasaki["Seller Type"].value_counts().plot(kind="bar")
plt.xticks(rotation=0)
plt.title("Dealer Generates the highest sales of Kawasaki bikes")
plt.show()
```

```python
plt.scatter(kawasaki["Price (INR)"],kawasaki["Resale Price (INR)"],color="green", s=7, alpha=0.7)
plt.xlabel("Sales Price")
plt.ylabel("Resale Price")
plt.title("Positive Correlation between Price Vs Resale Price")
plt.show()
```

```python
data=kawasaki.groupby("Model")["Mileage (km/l)"].sum().sort_values(ascending=False)
top = data.idxmax()
plt.plot(data.index.astype(str), data.values, marker="o", color="red", linewidth=3)
plt.title(f"The Kawasaki {top} model has the highest Mileage")
plt.xlabel("Model")
plt.ylabel("Mileage (km/l)")   
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
kawasaki["Insurance Status"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90 )
plt.title("Most Kawasaki bikes have expired insurance")
plt.show()
```
